```python
def simple_plot(fnames, grades, path_folder, size=(3,3), fs=12):
    fig = plt.figure(figsize=size)
    n = len(fnames)
    j = 0
    
    for i in range(n):
        j += 1
        ax = fig.add_subplot(1, n, j)
        img = plt.imread(os.path.join(path_folder, fnames[i]))
        ax.imshow(img)
        ax.set_title(f"{fnames[i].split('.')[0], grades[i]}", loc="center", fontsize=fs)
        ax.axis('off')
    plt.show()
```

```python
def move_dict_element(d, key_to_move, new_position):
    items = list(d.items())
    item = items.pop([k for k, v in items].index(key_to_move))
    items.insert(new_position, item)
    return dict(items)
```

```python
dic_patch_conf = {}
dic_np_val = {}

for model_name, dic_model_sens in dic_non_overlap_deletion_fundus_cres.items():
    avg_class_conf_drop = {}
    list_conf_drop = []
    #print('Model: ', model_name)
    
    for grade, class_sensitivity in dic_model_sens.items():
        avg_class_kdel_conf_drop = []
        #print('Grade: ', grade)
        
        for k, df_sensitiv in class_sensitivity.items():            
            #print('class_sensitivity: \n', class_sensitivity)
            #avg_class_kdel_conf_drop.append(df_sensitiv.conf_drop.mean())
            #avg_class_kdel_conf_drop.append(df_sensitiv.occ_pred.mean())  
            #print(k, avg_class_kdel_conf_drop)
            avg_class_kdel_conf_drop.append(df_sensitiv.relative_drop.mean())
        avg_class_conf_drop[grade] = avg_class_kdel_conf_drop
        list_conf_drop.append(avg_class_kdel_conf_drop)
        #break
        
    dic_patch_conf[model_name] = avg_class_conf_drop
    dic_np_val[model_name] = np.array(list_conf_drop) 
    #break
```
