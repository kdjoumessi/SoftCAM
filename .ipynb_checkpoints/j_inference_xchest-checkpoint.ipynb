```python
import os
import cv2
import torch
import pandas as pd
import matplotlib.pyplot as plt
from torch.distributions import Categorical

from utils.func import load_conf_file
from modules.builder import generate_model

from utils.posthoc.tools import load_image
```

```python
path = '/gpfs01/berens/data/data/DR/Kaggle_DR_resized'
img_path = '/gpfs01/berens/data/data/DR/Kaggle_DR_resized/kaggle_data_512'
```

```python
config_file = './configs/default.yaml'
cfg = load_conf_file(config_file)
cfg.train.checkpoint = None
cfg.train.network = 'bagnet33'
cfg.train.is_fpn = True
cfg.data.num_classes = 5 

cfg.data.mean = [0.41326871514320374, 0.2723627984523773, 0.18590997159481049]
cfg.data.std  = [0.29345420002937317, 0.20033970475196838, 0.15474912524223328]
```

```python
def plot_img_heat_att(imgs, dic, size, txt_kwargs):
    fig = plt.figure(figsize=size)
    n = len(imgs)
    j = 0
    
    for i, img in enumerate(imgs):
        j += 1
        ax = fig.add_subplot(1, n, j)
        ax.imshow(img)
        ax.set_title(dic[i], loc="center", **txt_kwargs)
        ax.axis('off')
    plt.show()
```

## Load data

```python
df = pd.read_csv(path + '/kaggle_gradable_test_new_qual_eval.csv')
df_disease = df[df.level > 0].reset_index(drop=True)
df.head(3)
```

## Load the model

```python
model_path = './model_weights/SA_best_validation_weights_acc.pt'

model = generate_model(cfg)
model.to('cuda')
model.load_state_dict(torch.load(model_path, weights_only=False), strict=True)
```

## Load the data

```python
sample = df_disease.sample(1)
fname = sample.filename.tolist()[0]
level = sample.level.tolist()[0]
level
```

```python
ts_img, np_img = load_image(cfg, img_path, fname)
ts_img = ts_img.to('cuda')
```

```python
prediction, activation, att_weight = model(ts_img)
prediction = prediction.data.cpu()
activation = activation[0].detach().cpu().numpy()
att_weight = att_weight[0].detach().cpu().numpy()
```

```python
dist = Categorical(logits=prediction)
pred, yhat = torch.topk(dist.probs, 1) 

print('Logit distribution: \t ', prediction.numpy()[0])
print(f'Probabilities: \t\t {round(pred.item(), 4)}, \t Class: {yhat.item()}')
```

```python
s = 512
act = cv2.resize(activation[level], dsize=(s,s), interpolation=cv2.INTER_CUBIC)
att = cv2.resize(att_weight[0], dsize=(s,s), interpolation=cv2.INTER_CUBIC)
```

```python
size=(10,12) 
ll = [np_img, act, att]
txt_kwargs = dict(color="black", fontsize=14)
lll = {0: f"Grade: {level}", 1:'Heatmap', 2:'Attention'}

plot_img_heat_att(ll, lll, size, txt_kwargs)
```

```python
ll = [np_img, act, att]
plot_img_heat_att(ll, lll, size, txt_kwargs)
```
